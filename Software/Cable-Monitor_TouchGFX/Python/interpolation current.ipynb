```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# s-Werte
s_mag = np.array([2162, 1195, 477], dtype=np.float64)

# d-Werte
d = np.array([10, 50, 100], dtype=np.float64)

# s-Werte
s = np.array([1304, 924, 695], dtype=np.float64)

# Koeffizienten der Parabel
a = ((s[0]**2) * (s[1] - s[2]) * d[0] - s[0] * ((s[1]**2) * d[1] - (s[2]**2) * d[2]) + s[1] * (s[1] * d[1] - s[2] * d[2]) * s[2]) / ((s[0]**2 - s[0] * (s[1] + s[2]) + s[1] * s[2]) * (s[1] - s[2]))

b = -(s[0]**2 * (s[1]**2 * (d[0] - d[1]) - s[2]**2 * (d[0] - d[2])) + s[1]**2 * s[2]**2 * (d[1] - d[2])) / ((s[0]**2 - s[0] * (s[1] + s[2]) + s[1] * s[2]) * (s[1] - s[2]))

c = (s[0] * (s[0] * (s[1] * (d[0] - d[1]) - s[2] * (d[0] - d[2])) + s[1] * s[2] * (d[1] - d[2])) * s[1] * s[2]) / ((s[0]**2 - s[0] * (s[1] + s[2]) + s[1] * s[2]) * (s[1] - s[2]))

def func(x):
    return a + (b/x) + (c/x**2)

# x-Werte für die Parabel
y = np.linspace(100, 500, 1000)

# y-Werte für die Parabel
x = func(y)

# Plot der Parabel
plt.plot(x, y, label='Parabel')
plt.xlabel('Distance in mm')
plt.ylabel('Signal Strength')
plt.grid()
plt.show()

a_magn1 = 5 / (s_mag[0] * d[0])
a_magn2 = 5 / (s_mag[1] * d[1])
a_magn3 = 5 / (s_mag[2] * d[2])

y = 2000;
x = func(1250)

# test if distance x is in range of sample d[0] and calculate current using coefficient a_magn1
if 0 <= x <= d[0] or d[0] < x <= d[1]/2:
    I = y * x * a_magn1
# test if distance x is in range of sample d[1] and calculate current using coefficient a_magn2
elif d[1]/2 < x <= d[1] or d[1] < x <= d[2]/2:
    I = y * x * a_magn2
# test if distance x is in range of sample d[2] and calculate current using coefficient a_magn3
elif d[2]/2 < x:
    I = y * x * a_magn3

print(I)
```
