```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# d-Werte
d = np.array([10, 50, 100], dtype=np.float64)

# s-Werte
s = np.array([485, 204, 142], dtype=np.float64)

# Koeffizienten der Parabel
a = ((s[0]**2) * (s[1] - s[2]) * d[0] - s[0] * ((s[1]**2) * d[1] - (s[2]**2) * d[2]) + s[1] * (s[1] * d[1] - s[2] * d[2]) * s[2]) / ((s[0]**2 - s[0] * (s[1] + s[2]) + s[1] * s[2]) * (s[1] - s[2]))

b = -(s[0]**2 * (s[1]**2 * (d[0] - d[1]) - s[2]**2 * (d[0] - d[2])) + s[1]**2 * s[2]**2 * (d[1] - d[2])) / ((s[0]**2 - s[0] * (s[1] + s[2]) + s[1] * s[2]) * (s[1] - s[2]))

c = (s[0] * (s[0] * (s[1] * (d[0] - d[1]) - s[2] * (d[0] - d[2])) + s[1] * s[2] * (d[1] - d[2])) * s[1] * s[2]) / ((s[0]**2 - s[0] * (s[1] + s[2]) + s[1] * s[2]) * (s[1] - s[2]))

def func(x):
    return a + (b/x) + (c/x**2)

# x-Werte für die Parabel
x = np.linspace(100, 500, 1000)


# y-Werte für die Parabel
y = func(x)

# Plot der Parabel
plt.plot(y, x, label='Parabel')
plt.xlabel('Distance in mm')
plt.ylabel('Signal Strength')
plt.grid()
plt.show()
```
